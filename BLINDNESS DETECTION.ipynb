```python
# This function will act as a filter for the image data

def load_colorfilter(image, sigmaX=10):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image=cv2.addWeighted ( image,4, cv2.GaussianBlur( image , (0,0) , sigmaX),-4 ,128)
    return image
```

```python
for i in range(len(images)):
    output = load_colorfilter(images[i])
    images[i] = output
```

```python
X = (np.array(images) /255.0)
y = dataset.diagnosis.values
X, y
```

```python
length_X = len(X)
length_y = len(y)

print("Length of X:", length_X)
print("Length of y:", length_y)
```

```python
sys.stdout.flush()
aug = ImageDataGenerator(rotation_range=0.2, width_shift_range=0.2, \
    height_shift_range=0.2, shear_range=0.2, zoom_range=0.2,\
    horizontal_flip=True, fill_mode="nearest")
```

```python

X, X_test, y, y_test = train_test_split(X,y,test_size=0.1, stratify = y)
X.shape, X_test.shape, y.shape, y_test.shape
```

```python
def display_training_curves(training, validation, title, subplot):
    
    if subplot%10==1: # set up the subplots on the first call
        plt.subplots(figsize=(10,10), facecolor='#F0F0F0')
        plt.tight_layout()
    ax = plt.subplot(subplot)
    ax.set_facecolor('#F8F8F8')
    ax.plot(training)
    ax.plot(validation)
    ax.set_title('model '+ title)
    ax.set_ylabel(title)
    #ax.set_ylim(0.28,1.05)
    ax.set_xlabel('epoch')
    ax.legend(['train', 'valid.'])
    plt.show()
```

```python
BS = 1000       #Batch size
accuracy = []
skf = StratifiedKFold(n_splits=5)
skf.get_n_splits(X,y)

for train, test in skf.split(X,y):
    model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), input_shape=(150, 150, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.BatchNormalization(),

    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.BatchNormalization(),

    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.BatchNormalization(),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),

    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.BatchNormalization(),

    tf.keras.layers.Dense(5, activation='softmax')
    
    ])
    initial_learning_rate = 0.001
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate, decay_steps=10000, decay_rate=0.9, staircase=True)
    optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule)

    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['acc'])

    history = model.fit_generator(aug.flow(X[train], y[train], batch_size=1000),
    validation_data=(X[test], y[test]),
    epochs=10, verbose = 1)
    acc=model.evaluate(X[test], y[test])
    accuracy.append(acc[1])
    display_training_curves(
    history.history['loss'], 
    history.history['val_loss'], 
    'loss', 211)
    
    display_training_curves(
    history.history['acc'], 
    history.history['val_acc'], 
    'accuracy', 212)
```
